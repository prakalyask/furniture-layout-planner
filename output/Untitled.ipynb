```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Get the current working directory
directory = os.getcwd()  # This will use the current working directory

# Optionally, you can append a subfolder if you want to look in a specific folder within the CWD
# For example, if you have an "output" folder inside your current directory:
# directory = os.path.join(os.getcwd(), 'output')

# Check if the directory exists
if not os.path.isdir(directory):
    print(f"Directory not found: {directory}")
else:
    # Define image file extensions to look for
    image_extensions = ['.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tiff']

    # List all files in the directory
    files = os.listdir(directory)

    # Filter files to only include images based on extension
    image_files = [file for file in files if any(file.lower().endswith(ext) for ext in image_extensions)]

    # Loop through and display each image
    for image_file in image_files:
        # Create full path to the image file
        image_path = os.path.join(directory, image_file)

        # Open the image
        img = Image.open(image_path)

        # Get the image size
        img_width, img_height = img.size

        # Display the image in original resolution (without resizing)
        plt.figure(figsize=(img_width / 100, img_height / 100))  # Adjust figure size to match the image resolution
        plt.imshow(img)
        plt.axis('off')  # Hide the axis
        plt.show()
```
